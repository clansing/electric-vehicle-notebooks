This notebook demonstrates how to read and plot data from the `Electric_Vehicle_Population_Data.csv` file using pandas and matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Read the CSV file
data_dir = os.environ.get('DATA_DIR')
print(f"Data dir = {data_dir}")

df = pd.read_csv(os.path.join(data_dir, 'Electric_Vehicle_Population_Data.csv'))
df.head()
print('Columns:', df.columns.tolist())
```

```python
# 1) count vehicles per postal code
counts = df.groupby('Postal Code').size().sort_values(ascending=False)

# 2) plot
plt.figure(figsize=(12, 6))
counts.plot(kind='bar')
plt.title('Total Vehicles per Postal Code')
plt.xlabel('Postal Code')
plt.ylabel('Number of Vehicles')
plt.tight_layout()
plt.show()
```
